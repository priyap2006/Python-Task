```python
import pandas as pd
df = pd.read_csv("weather_data.csv")
df
```

```python
new_df = df.fillna(6)
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```
