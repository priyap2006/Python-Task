```python
1.Maximum Subarray
```

```python
def maxSubArray(nums):
    curr_sum = nums[0]
    max_sum = nums[0]
    for num in nums[1:]:
        curr_sum = max(num, curr_sum + num)
        max_sum = max(max_sum, curr_sum)
    return max_sum

nums = list(map(int, input("Enter array: ").split()))
print("Output:", maxSubArray(nums))
```

```python
2. Single Number
```

```python
def singleNumber(nums):
    res = 0
    for n in nums:
        res ^= n
    return res

nums = list(map(int, input("Enter array: ").split()))
print("Output:", singleNumber(nums))
```

```python
3.Valid Parentheses
```

```python
def isValid(s):
    stack = []
    mapping = {')':'(', '}':'{', ']':'['}
    for c in s:
        if c in mapping:
            if not stack or stack.pop() != mapping[c]:
                return False
        else:
            stack.append(c)
    return not stack

s = input("Enter string: ")
print("Output:", isValid(s))
```

```python
4.Move Zeroes
```

```python
def moveZeroes(nums):
    j = 0
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[i], nums[j] = nums[j], nums[i]
            j += 1

nums = list(map(int, input("Enter array: ").split()))
moveZeroes(nums)
print("Output:", nums)
```

```python
5.Valid Anagram
```

```python
from collections import Counter

def isAnagram(s, t):
    return Counter(s) == Counter(t)

s = input("Enter string s: ")
t = input("Enter string t: ")
print("Output:", isAnagram(s, t))
```

```python
6.Implement strStr()
```

```python
def strStr(haystack, needle):
    if needle == "":
        return 0
    for i in range(len(haystack) - len(needle) + 1):
        if haystack[i:i+len(needle)] == needle:
            return i
    return -1

haystack = input("Enter haystack: ")
needle = input("Enter needle: ")
print("Output:", strStr(haystack, needle))
```

```python
7.Excel Sheet Column Number
```

```python
def titleToNumber(columnTitle):
    res = 0
    for c in columnTitle:
        res = res * 26 + (ord(c) - ord('A') + 1)
    return res

col = input("Enter column title: ")
print("Output:", titleToNumber(col))
```

```python
9.Reverse Integer
```

```python
def reverse(x):
    sign = -1 if x < 0 else 1
    x *= sign
    rev = 0
    while x > 0:
        rev = rev * 10 + x % 10
        x //= 10
    rev *= sign
    return rev if -2**31 <= rev <= 2**31-1 else 0

x = int(input("Enter integer: "))
print("Output:", reverse(x))
```

```python
10.Maximum Product of Three Numbers
```

```python
def maximumProduct(nums):
    nums.sort()
    return max(nums[-1]*nums[-2]*nums[-3],
               nums[0]*nums[1]*nums[-1])

nums = list(map(int, input("Enter array: ").split()))
print("Output:", maximumProduct(nums))
```

```python
11.Majority Element
```

```python
def majorityElement(nums):
    count = 0
    candidate = None
    for num in nums:
        if count == 0:
            candidate = num
        count += 1 if num == candidate else -1
    return candidate

nums = list(map(int, input("Enter array: ").split()))
print("Output:", majorityElement(nums))
```

```python
12.Happy Number
```

```python
def isHappy(n):
    seen = set()
    while n != 1:
        if n in seen:
            return False
        seen.add(n)
        n = sum(int(d)**2 for d in str(n))
    return True

n = int(input("Enter number: "))
print("Output:", isHappy(n))
```

```python
13.Power of Three
```

```python
def isPowerOfThree(n):
    if n < 1:
        return False
    while n % 3 == 0:
        n //= 3
    return n == 1

n = int(input("Enter number: "))
print("Output:", isPowerOfThree(n))
```

```python
14.Power of Four
```

```python
def isPowerOfFour(n):
    if n < 1:
        return False
    while n % 4 == 0:
        n //= 4
    return n == 1

n = int(input("Enter number: "))
print("Output:", isPowerOfFour(n))
```

```python
15.Factorial Trailing Zeroes
```

```python
def trailingZeroes(n):
    count = 0
    while n > 0:
        n //= 5
        count += n
    return count

n = int(input("Enter number: "))
print("Output:", trailingZeroes(n))
```

```python
16.Number of 1 Bits
```

```python
def hammingWeight(n):
    count = 0
    while n:
        n &= n - 1
        count += 1
    return count

n = int(input("Enter number: "))
print("Output:", hammingWeight(n))
```

```python
17.Missing Number
```

```python
def missingNumber(nums):
    n = len(nums)
    return n*(n+1)//2 - sum(nums)

nums = list(map(int, input("Enter array: ").split()))
print("Output:", missingNumber(nums))
```

```python
18.Intersection of Two Arrays
```

```python
def intersection(nums1, nums2):
    return list(set(nums1) & set(nums2))

nums1 = list(map(int, input("Enter first array: ").split()))
nums2 = list(map(int, input("Enter second array: ").split()))
print("Output:", intersection(nums1, nums2))
```

```python
20.Fizz Buzz
```

```python
def fizzBuzz(n):
    result = []
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            result.append("FizzBuzz")
        elif i % 3 == 0:
            result.append("Fizz")
        elif i % 5 == 0:
            result.append("Buzz")
        else:
            result.append(str(i))
    return result

n = int(input("Enter a number: "))
print("Output:", fizzBuzz(n))
```
